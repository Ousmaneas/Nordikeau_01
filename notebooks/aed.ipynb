```python
import pandas as pd
import os
```

```python
def load_data(data_type, db_name, tb_name, variables):
    
    # Définition du chemin de sauvegarde
    data_path = r"C:\Users\ymarega\NORDIKeau\INO-PartenariatNordikeauINRSeau - INO-25-0001-Projet[Nordikeau-INRSeau]\Data"    

    ## Chemin complet du fichier
    data_type_db_tb_name_path  = os.path.join(data_path, data_type, db_name, tb_name)
    print(data_type_db_tb_name_path)

    # Lire le fichier CSV et trier par date_creation
    data_type_db_tb_name = pd.read_csv(data_type_db_tb_name_path, usecols=variables)
    data_type_db_tb_name = data_type_db_tb_name.sort_values(by="date_creation").reset_index(drop=True)

    return data_type_db_tb_name
    
```

```python
# type de données
data_type = "raw"

## Nom de la base de donnée     
db_name = 'saint_esprit'

## Nom de la table
tb_name = 'stesp_p_usine_fl_20190808_20210115.csv'

## variable 
variables = ['date_creation'] 

df = load_data(data_type, db_name, tb_name, variables)
df
```
